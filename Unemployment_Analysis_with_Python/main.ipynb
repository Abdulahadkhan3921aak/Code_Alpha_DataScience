```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Unemployment in India.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
print(df.columns.tolist())

df.columns = df.columns.str.strip()

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
# collect the date range
start_date = df['Date'].min()
end_date = df['Date'].max()
print(f"Date range: {start_date} to {end_date}")
```

```python
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
# collect the Regions in the region column
regions = df['Region'].unique()
print(f"Regions: {regions}")
```

```python
Urbandf = df[df['Area'] == 'Urban']
Ruraldf = df[df['Area'] == 'Rural']
```

```python
for region in regions:
    plt.figure(figsize=(12, 6))
    plt.title(f"Unemployment Rate in {region} (Urban)")
    plt.axvline(pd.to_datetime('2020-03-01'), color='r', linestyle='--', label='COVID-19 Start')
    sns.lineplot(data=Urbandf[Urbandf['Region'] == region],
                 x='Date', y='Estimated Unemployment Rate (%)', marker='o')
    plt.xticks(rotation=45)
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.tight_layout()
    plt.grid()
    plt.show()
```

```python
for region in regions:
    plt.figure(figsize=(12, 6))
    plt.title(f"Unemployment Rate in {region} (Rural)")
    plt.axvline(pd.to_datetime('2020-03-01'), color='r', linestyle='--', label='COVID-19 Start')
    sns.lineplot(data=Ruraldf[Ruraldf['Region'] == region],
                 x='Date', y='Estimated Unemployment Rate (%)', marker='o')
    plt.xticks(rotation=45)
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.tight_layout()
    plt.grid()
    plt.show()
```
